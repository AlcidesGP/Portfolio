Neste exerccio vamos trabalhar com a base "Heart Disease Data Set" disponvel no site do [UCI](https://archive.ics.uci.edu/ml/datasets/heart+disease).

Temos 303 indivduos e vamos utilizar 13 variveis disponveis para prever a presena de doena cardaca na varivel ```num```(a 14$^a$ varivel).

A descrio das variveis est recortada abaixo:
- age: idade do paciente em anos
- sex: sexo (1 = male; 0 = female)  
- cp: tipo de dor no peito
  - 1: angina tpica
  - 2: angina atpica
  - 3: dor no-angina
  - 4: assintomtico
- trestbps: presso sangunea em repouso (em mm Hg na admisso ao hospital
- chol: colesterol srico em mg/dl
- fbs: (acar no sangue em jejum > 120 mg/dl) (1 = True; 0 = False)
- restecg: resultados eletrocardiogrficos em repouso
  - 0: normal
  - 1: tendo anormalidade da onda ST-T (Inverses de onda T e / ou ST com elevao ou depresso de > 0.05 mV)
  - 2: mostrando hipertrofia ventricular esquerda provvel ou definitiva pelos critrios de Estes
- thalach: frequncia cardaca mxima alcanada
- exang: angina induzida por exerccio(1 = sim; 0 = no)
- oldpeak = Depresso de ST induzida por exerccio em relao ao repouso
- slope: Depresso de ST induzida por exerccio em relao ao repouso
  - 1: inclinao ascendente
  - 2: estvel
  - 3: inclinao descendente
- ca: nmero de vasos principais (0-3) coloridos por fluorosopia
- thal: 3 = normal; 6 = defeito corrigido; 7 = defeito reversvel
- num: diagnstico de doena cardaga (status de doena angiogrfica)

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.formula.api as smf
````

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

df = pd.read_csv(url, 
                 names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                        'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'])
df.head()
````

````python
df['flag_doente'] = (df['num'] != 0).astype('int64')
df
````

1. Monte uma bivariada para a varivel ```sex``` com as seguintes estatsticas:

- Nmero de *saudveis*
- Nmero de *doentes*
- Total
- Quantidade mdia de doentes
- *Odds*
- *Odds ratio* (Use o total como referncia)
- Logito (LOG(*Odds*))
- O LOG do *Odds ratio* (com o total como referncia)  chamado de *Weight of Evidence* (WOE). Calcule-o.

````python
def transform(x):
    if x == 0.0:
        return 'mulher'
    else:
        return 'homem'
    
def analise():
    # # Dummies de doentes e saudaveis
    df_analise = pd.get_dummies(df,columns=['flag_doente'])
    # Renomear as dummies
    df_analise = df_analise.rename(columns ={'flag_doente_0':'saudavel','flag_doente_1':'doente'})
    # Nomear os nomes do atributo sexo
    df_analise['sex'] = df_analise['sex'].apply(lambda x: transform(x)) 
    display(df_analise.head())
    return df_analise

def agrupamento(data):
    dt = data.copy()
    # Bivariada com base no sexo para pacientes saudaveis e doentes.
    dt = dt.groupby('sex')[['saudavel','doente']].sum()
    # Total de casos
    dt['Total'] = dt.sum(1)
    # Mdia dos doentes
    dt[['media_saudavel','media_doente']] =  data.groupby('sex')[['saudavel','doente']].mean()
    # ODDs
    dt['chance'] = dt['saudavel'] / dt['doente']
    # Taxa de ODDs
    dt['taxa_chance'] = dt['Total'] / dt['chance']
    # Logito
    dt['log_chance'] = np.log(dt['chance'])
    # Weight of Evidence
    dt['WOE'] = np.log(dt['taxa_chance'])
    
    return dt
````

````python
df_analise = analise()
````

````python
agrupamento(df_analise)
````

2. Monte uma bivariada para a varivel ```age``` com as seguintes estatsticas:
- Categorize a varivel em 5 grupos
- Calcule as mesmas quantidades calculadas no execcio anterior
- Calcule tambm o valor mdio da varivel ```age```

````python
def bivariada_age(data):
    dt = data.copy()
    dat = data.copy()
    dt['cut_age'] = pd.qcut(dt['age'], q=5).astype(str)
    dat['cut_age'] = dt['cut_age'].copy()
    
    
    dt = dt.groupby('cut_age')[['saudavel','doente']].sum()
    # Total de casos
    dt['Total'] = dt.sum(1)
    # Mdia dos doentes
    dt[['media_saudavel','media_doente']] =  dat.groupby('cut_age')[['saudavel','doente']].mean()
    # ODDs para pessoa ficar doente.
    dt['chance'] = dt['doente'] /dt['saudavel'] 
    # Taxa de ODDs
    dt['taxa_chance'] = dt['Total'] / dt['chance']
    # Logito
    dt['log_chance'] = np.log(dt['chance'])
    # Weight of Evidence
    dt['WOE'] = np.log(dt['taxa_chance'])
    # Mdia de idade
    dt['media'] = dat.groupby('cut_age')['age'].mean()
    return dt
    
````

````python
bivariada_age(df_analise)
````

3. Visualize a informao da bivariada construda acima atravs de um grfico da sua escolha.

````python
def grafico_infos():
    dt = (bivariada_age(df_analise)
         .reset_index())
    #sns.despine()
    df_an = pd.melt(dt, id_vars='cut_age',value_vars=['saudavel','doente'])
    figura = plt.figure(figsize=(15,30))
    plt.suptitle("Anlise grfica.", fontsize=30)
    plt.subplots_adjust(top=0.94,
                        hspace = 0.5)
    tamanho = (6,1)
    pos = (0,0)
    with sns.axes_style("darkgrid"):
        ax = plt.subplot2grid(tamanho, pos)
    
    fig = sns.barplot(data=df_an, y='value',x='cut_age', hue='variable', ax=ax)
    fig.set_title('Relao entre Doentes e Saudaveis para faixa de idade',fontsize=20)
    
    lista = ['chance','taxa_chance','log_chance','WOE','media_doente']
    
    for i in range(5):
        
        pos = (i+1,0)
        with sns.axes_style("darkgrid"):
            ax = plt.subplot2grid(tamanho, pos)
        
        fig = sns.lineplot(data=dt, y=lista[i], x='cut_age',ax=ax)
        fig.set_title(f'Relao entre {lista[i]} para faixa de idade',fontsize=20)
````

````python
grafico_infos()
````
