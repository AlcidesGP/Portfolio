```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec

df = pd.read_csv("C:\\Users\\alcid\\GitHub\\Portfolio\\EBAC\\Dados\\SINASC_RO_2019.csv")
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.
####  2) Faça um boxplot do peso do bebê
####  3) Faça um histograma da idade da mãe
#### 4) Faça um histograma do peso do bebê

```python
def dispersao_quant():
    dt = df.copy()
    figura = plt.figure(figsize=(15,6))
    outer = gridspec.GridSpec(1,2, wspace=0.3, hspace=0.5) # Tamanho de imagem.
    colunas = ['IDADEMAE','PESO']
    figura.suptitle('Dispersão dos dados quantitativos', fontsize=30, y=1.05)
    for co in range(2):
        inner = gridspec.GridSpecFromSubplotSpec(2, 1, subplot_spec=outer[co],wspace=0.1, hspace=0)
        coluna = colunas[co]
        # -------- Parte superior
        ax = plt.Subplot(figura, inner[0]) # Axis .
        figura.add_subplot(ax) # Adicionar o axis na figura.
                # Gráfico de box
        flierprops = dict(marker='x', markerfacecolor='r', markersize=4,
                  linestyle='none', markeredgecolor='orangered') 
        valores = df[coluna]

        box = sns.boxplot(valores, orient='h',width=.3,flierprops=flierprops, ax=ax)

        box.set_title(f'{coluna}',horizontalalignment='left', x=0,fontsize=15,color='dimgrey') # Título.
        
        quantiles = dt[coluna].quantile([0.25, 0.75]).values.tolist() # Separação dos quantiles.
        LI = quantiles[0] - ((quantiles[1] - quantiles[0])*1.5) # Definir limite inferior
        LS = quantiles[1] + ((quantiles[1] - quantiles[0])*1.5) # Definir limte superior
        if LI < dt[coluna].min(): # Loop que vai reposicionar em caso de valores abaixo/acima do cálculado.
            LI = dt[coluna].min()
        if LS > dt[coluna].max():
            LS = dt[coluna].max()

        posse = ['right', 'left'] # Reajustar a posição no boxplot.
        for ix in [int(LI),int(LS)]: # Plotar no gráfico.
            box.annotate(round(ix,2), xy=(ix, 0.18), fontsize=10, color='black')
        for ix,poss in zip(quantiles,posse):
            box.annotate(round(ix,2), xy=(ix, -0.18), fontsize=10, color='black',horizontalalignment=poss)

        box.annotate(round(dt[coluna].median(),2), xy=(dt[coluna].median(), 0.22),                     
                     fontsize=10, color='black',horizontalalignment='center')

        box.axis('off') # Desativar barras de eixo.
        box.tick_params(axis='x', colors='darkgrey')
        box.set_xlim([0,max(valores)*1.1])
        
        
        
        # -------- Parte inferior
        ax = plt.Subplot(figura, inner[1]) # Axis .
        figura.add_subplot(ax) # Adicionar o axis na figura.
        
        hist = sns.histplot(data=dt, x = coluna,color='blueviolet', stat='frequency',kde=True) # Plotagem Histograma.
                          # Desativar
        hist.spines['top'].set_visible(False)
        hist.spines['right'].set_visible(False)

                         # Alterar
        hist.spines['bottom'].set_color('darkgrey')
        hist.tick_params(axis='y', colors='black')
        hist.tick_params(axis='x', colors='darkgrey')
        hist.set(xlabel='Intervalo de variação', ylabel='')
        hist.set_xlim([0,max(valores)*1.1])
        
    return plt.show()
```

```python
dispersao_quant()
```
