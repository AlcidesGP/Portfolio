# Mtodos hierrquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

import scipy.cluster.hierarchy as shc
```

<a id=1></a>
## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variveis para mdia zero e desvio padro 1
4. Faa um agrupamento hierrquico dividindo essa base em 3 grupos
5. Faa o Dendrograma

```python
# Carregar o conjunto de dados dos pinguins
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Averguar Vazios
penguins.isnull().sum().sum()
```

```python
# Tratar vazios
df = penguins.dropna()
```

```python
# Verificar se h duplicados
df.duplicated().sum()
```

```python
def cluster(dt) -> None:
    # Copiando base de dados, selecionando quantitativos e padronizando-os.
    original = dt.copy()
    data = dt.copy()
    data = data.select_dtypes(include=np.number)
    data.iloc[:,:] = StandardScaler().fit_transform(data)
    
    # Criando modelo agloremativo.
    cluster = AgglomerativeClustering(n_clusters=3, linkage='complete')
    cluster.fit(data)
    data['grupos'] = cluster.labels_
    original['grupos'] = cluster.labels_
    
    # Criando Dendograma.
    shc.dendrogram(shc.linkage(data, method='complete'))
    
    
    # Retornando Grupos ao dado original 
    return original

analise = cluster(df)
```

## 2) Avalie os grupos

Descreva os grupos:

- Quantos pinguins de cada espcie h em cada grupo?
- Descreva as variveis quantitativas de cada grupo com as tcnicas que voc preferir.

```python
def grup_aval():
    print(f"\n{' '*15}Grupos x Especies")
    display(analise.groupby(['grupos','species'])[['species']].count())
    print(f"\n{' '*10}Mdia das variveis quantitativas\n")
    display(analise.groupby('grupos')[analise.select_dtypes(include=np.number).columns.tolist()].mean())
    
    
grup_aval()
```

```python
def cluster(dt) -> None:
    # Copiando base de dados, selecionando quantitativos e padronizando-os.
    original = dt.copy()
    data = dt.copy()
    data = data.select_dtypes(include=np.number)
    data.iloc[:,:] = StandardScaler().fit_transform(data)
    
    # Criando modelo agloremativo.
    cluster = AgglomerativeClustering(n_clusters=5, linkage='complete')
    cluster.fit(data)
    data['grupos'] = cluster.labels_
    original['grupos'] = cluster.labels_
    
    # Criando Dendograma.
    shc.dendrogram(shc.linkage(data, method='complete'))
    print(f"\n{' '*15}Grupos x Especies")
    display(original.groupby(['grupos','species'])[['species']].count())
    print(f"\n{' '*10}Mdia das variveis quantitativas\n")
    display(original.groupby('grupos')[original.select_dtypes(include=np.number).columns.tolist()].mean())
    
    # Retornando Grupos ao dado original 
    return original

analise = cluster(df)
```
