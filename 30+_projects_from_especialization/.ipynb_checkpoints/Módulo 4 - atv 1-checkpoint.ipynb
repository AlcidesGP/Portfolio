#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na tempreatura global, A NASA disponibiliza dados de diferenças de de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
import pandas as pd
import numpy as np

# Seu código aqui
""" Criar um dicionário contendo as informações para então passar para o formato DataFrame do pandas."""
dados = {'ano':[1900,1920,1940,1960, 1980,2000,2020],
         'anomalia termica': [-0.08,-.27,.12,-.03,.26,.40,1.02]}

df = pd.DataFrame(dados)
display(df)
```

#### 2) Coloque os anos nos índices conforme a tabela.

```python

# seu código aqui
df = df.set_index('ano') # Usar a função set_index para a tabela entre () poder ser o índice.
display(df)
```

#### 3) A partir do dicionário abaixo, crie uma séries do Pandas:

```python
dic_temperaturas = {1900: -.08, 1920: -.27, 1940: .12, 1960: -.03, 1980: .26, 2000: .40, 2020: 1.02}

# seu código aqui
df_2 = pd.Series(dic_temperaturas)
display(df_2)
```

#### 4) Transforme o ndarray abaixo em um dataframe. 
O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

```python
arr = np.random.normal(100, 10, (20,3))

# seu código aqui
df_3 = pd.DataFrame(arr)
display(df_3)
```

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#seu código aqui
df_3 = df_3.rename(columns=({0:'X1',1:'X2',2:'X3'})) # Usar a função rename para renomear as colunas
df_3.index = range(1,21,1) # Usando a chamada index para alterar o nome dos índices. Novos nomes criados usando um range de 1 a 21.
display(df_3)
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui
df_3['mean'] = df_3.mean(1)
display(df_3)
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui
df_3['log_mean'] = np.log2(df_3['mean']) # Usando a função do numpy "log2" para realizar a operação.
display(df_3)
```
