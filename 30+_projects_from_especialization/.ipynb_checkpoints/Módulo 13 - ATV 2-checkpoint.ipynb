# Avaliaçã com LASSO  e RIDGE

```python
def model(data, alvo, reg_model) -> None:
    """
    --------- Entrada -------
    data -> DataFrame com os atributos independentes
    alvo -> O atributo preditor
    reg_model -> Qual modelo de regressão linear, sendo lasso ou ridge
    
    
    ---------  Explicação do modelo ----------
    
    1. Atributos
        Selecionar quais modelos são categóricos e numéricos para serem colocados no preditor do statsmodels OLS.
    2. Criação do alphas e uma lista de resultados
    3. Definir L1.
        Se o modelo for Lasso o L1 será igual à 1. Caso seja Ridge o L1 será igual a 0.0001, pois se for zero a função
        summary fica desativada.
    4. Loop que vai variar em cada alpha.
        4.1 Separação de treino e teste
        4.2 Definir X e y
        4.3 Treinamento e análise de resultado
            Aqui será passado só a base de treino no modelo OLS e será extraido as informações de AIC, r2 e adjusted r2
            para o alpha do momento.
        4.4 Transformação
            Devido ao erro gerado para previsão com o OLS em base de treinamento dividida, a biblioteca SKLEARN vai ser 
            utilizada para as previsões e para isso é necessário que os dados estejam em formato numérico.
        4.5 predição no SKLEARN
            Os dados são novamentes separados (mantendo o mesmo random_state para a divisão de treino e teste ser mesma)
            e o modelo treinado gera a previsão que passa pelo R2 de teste para previsão.
        4.6 Salvar os resultados
            A lista é então salva
    5. É retornado um DataFrame contendo os resultados para cada alpha.
    
    """
    
    
    
    # 1. Atributos
    numericos = 'idade + tempo_emprego + C(qtd_filhos, Treatment(0)) + C(qt_pessoas_residencia, Treatment(2))'
    categoricos = 'C(data_ref, Treatment(0)) + C(sexo, Treatment(0)) +  C(posse_de_veiculo, Treatment(0))'
    categoricos += ' + C(posse_de_imovel, Treatment(1)) + C(tipo_renda, Treatment(0)) + C(educacao, Treatment(2))'
    categoricos += '+ C(estado_civil, Treatment(0)) + C(tipo_residencia, Treatment(1)) + C(mau, Treatment(0))'
    atributos = f'renda ~ {numericos} + {categoricos}'
    
    # 2. 
    resultados = []
    alphas = [0, 0.001, 0.005, 0.01, 0.05, 0.1] 
    
    # 3. 
    if reg_model == 'ridge':
        L1 = 0.0001 # caso o valor seja 0, o modelo não retorna um summary
    else:
        L1 = 1 # Retonar uma funçaõ lasso.1
        
    # 4. 
    for alpha in alphas:
        # 4.1. Separação de treino e teste
        df_train, df_test, y_train, y_test = train_test_split(data, np.log(alvo), test_size=0.25, random_state=150)
        df_train['renda'] = y_train
        
        # 4.2.
        y, X = patsy.dmatrices(atributos, df_train)
        
        # 4.3. Treinamento e análise de resultado
        model = sm.OLS(y, X).fit_regularized(method='elastic_net', refit=True, L1_wt = L1, alpha = alpha)
        resultado = pd.DataFrame(model.summary().tables[0])
        resultado[3] = resultado[3].astype(str)
        aic = float(resultado.iloc[5,3])
        r2_sm = float(resultado.iloc[0,3].strip())
        adj_r2 = float(resultado.iloc[1,3].strip())
        dt = data.copy()
        #4.4. Transformação
        for alterar in dt.select_dtypes(exclude=np.number):
            dt[alterar] = LabelEncoder().fit_transform(dt[alterar])

        #4.5. Predição no SKLEARN
        df_train, df_test, y_train, y_test = train_test_split(dt, np.log(alvo), test_size=0.25, random_state=150)
        if reg_model == 'ridge':
            modelo = Ridge(alpha=alpha).fit(df_train, y_train) # Treino
        else:
            modelo = Lasso(alpha=alpha).fit(df_train, y_train) # Treino
        predict = modelo.predict(df_test) # Previsão 
        r2_prev = r2_score(y_test, predict) # Resultado
        
        #4.6 Salvar os resultados
        resultados.append([alpha, aic, r2_sm, adj_r2, r2_prev])
    
    #5.
    pd.set_option('display.float_format', '{:.7}'.format) # Exibir formato float
    return pd.DataFrame(resultados, columns=['alpha', 'aic', 'r2_sm', 'adj_r2', 'r2_prev'])
```

# Avaliação com stepwise e melhoria de modelo.
- Para essa etapa, vou utilizar o que foi aprendido no exercício passado, no qual o p-value apontou que as classes: data_ref, mau, posse_de_veiculo e estado_civil podem ser desconsiderados para a análise, que só irão contribuir para a complexidade sem acrescentar muita informação nova.

```python
def model_stepwise(data, alvo) -> None:
    
    
    # Atributos
    numericos = 'tempo_emprego + C(qtd_filhos, Treatment(0)) + C(qt_pessoas_residencia, Treatment(2))'
    categoricos = 'C(sexo, Treatment(0)) + C(tipo_residencia, Treatment(1))'
    categoricos += ' + C(posse_de_imovel, Treatment(1)) + C(tipo_renda, Treatment(0)) + C(educacao, Treatment(2))'
    atributos = f'np.log(renda) ~ {numericos} + {categoricos}'
    
    atributos = f'{numericos} + {categoricos}'
    
    atributos = atributos.split('+')
    
    resultados = []
    
    alvo = np.log(alvo)
    
    atributo_base = 'renda ~ idade '
    base = 'idade'
    at_base = ['idade']
    resultados = list()
    for atributo in atributos:
        
        df_train, df_test, y_train, y_test = train_test_split(data, np.log(alvo), test_size=0.25, random_state=100)
        df_train['renda'] = y_train
        atributo_base += f'+ {atributo}'

        y, X = patsy.dmatrices(atributo_base, df_train)
        
        # Treinamento e análise de resultado
        model = sm.OLS(y, X).fit_regularized(method='elastic_net', refit=True, L1_wt = 1, alpha = 0)
        resultado = pd.DataFrame(model.summary().tables[0])
        resultado[3] = resultado[3].astype(str)
        aic = float(resultado.iloc[5,3])
        r2_sm = float(resultado.iloc[0,3].strip())
        adj_r2 = float(resultado.iloc[1,3].strip())
        
        # Transformação
        dt = data.copy()
        

        try:
            atr = re.findall(r'\(.*\,',atributo)[0].replace('(','').replace(',','')
        except:
            atr = atributo
            
        at_base.append(atr.strip())
        dt = dt.loc[:,at_base]
        
        for alterar in dt.select_dtypes(exclude=np.number):
            dt[alterar] = LabelEncoder().fit_transform(dt[alterar])
            
        # Predição no SKLEARN
        df_train, df_test, y_train, y_test = train_test_split(dt, alvo, test_size=0.25, random_state=100)
        modelo = Lasso(alpha=0).fit(df_train, y_train) # Treino
        predict = modelo.predict(df_test) # Previsão 
        r2_prev = r2_score(y_test, predict) # Resultado
        base += ' '+atr
        # Salvar os resultados
        resultados.append([base, aic, r2_sm, adj_r2, r2_prev])      
           
    pd.set_option('display.float_format', '{:.7}'.format) # Exibir formato float
    return pd.DataFrame(resultados, columns=['atributos','aic', 'r2_sm', 'adj_r2', 'r2_prev'])
    
```

```python
def renda_analise(data, alvo):
    """
    ----- Entrada ------
    data -> DataFrame contendo os atributos independentes
    alvo -> Series contendo o atributo dependente.
    
    ----- Objetivo -------
    Gerar um gráfico que permite avaliar a distribuição de cada atributo para a criação de novos atributos preditores.
    
    ----- Funcionamento -------
    1. Seleção de atributos
        Com base no exercício anterior
        
    2. Binning
        Cortar um dado numérico em quartís, garantindo a proporção de 4 pedaços iguais, isso irá ajudar na análise 
        de distribuição 
        
    3. Gráfico
        3.1. Gráfico de caixa
        3.2. Gráfico de barra
    """
    
    
    # 1. Seleção de atributos
    atributos = 'idade tempo_emprego qtd_filhos qt_pessoas_residencia sexo tipo_residencia posse_de_imovel tipo_renda educacao'
    atributos = atributos.split(' ')
    
    dt = data.copy() # Cópia do dataframe
    dt = dt.loc[:,atributos] # Seleção de elementos
    
    # 2. Binning
    dt['renda_cat'] = pd.qcut(alvo,q=4 ).astype(str) 
    
    # 3. Gráfico
    figura = plt.figure(figsize=(15,70)) # Tamanho
    figura.suptitle('Análise das variáveis em relação à renda', fontsize=29) # Título
    figura.subplots_adjust(top=0.95, # Espaço entre o título e os gráficos
                            hspace = 0.5) # Distância entre gráficos.
    tamanho = (9,1) # Tamanho
    
    
    # 3.1. Gráfico de caixa
    for i in range(2): 
        
        pos = (i,0) # Posição na figura
        ax = plt.subplot2grid(tamanho, pos) # axis
        
        sns.boxplot(data=dt, x = atributos[i], y='renda_cat', orient='h',ax = ax) # plot
        
    # 3.2. Gráfico de barra 
    
    for i in range(tamanho[0]-2):
        pos = (i+2,0)
        ax = plt.subplot2grid(tamanho, pos)
    
    
        sns.countplot(data=dt, x=atributos[i+2],hue='renda_cat',ax =ax)
    
    return plt.show()


renda_analise(df, y_alvo)
```

```python
def renda_analise_gerado(data, alvo):
    """
    ----- Entrada ------
    data -> DataFrame contendo os atributos independentes
    alvo -> Series contendo o atributo dependente.
    
    ----- Objetivo -------
    Com base no gráfico anterior, criar novos dados que permitam melhorar as previsões dos dados e com essa função seja
    possível visualizar a dispersão desses novo atributos.
    
    ----- Funcionamento -------
    1. Seleção de atributos
        Com base no exercício anterior
        
    2. Binning
        Cortar um dado numérico em quartís, garantindo a proporção de 4 pedaços iguais, isso irá ajudar na análise 
        de distribuição 
        
    3. Criação de novos dados.      
    
    4. Gráfico.
    """
    # 1. Seleção de atributos
    atributos = 'idade tempo_emprego qtd_filhos qt_pessoas_residencia sexo tipo_residencia posse_de_imovel tipo_renda educacao'
    atributos = atributos.split(' ')

    
    dt = data.copy() # Criar cópia do dataframe
    dt = dt.loc[:,atributos] # Selecionar dados do dataframe
    
    
    # 2. Binning
    dt['renda_cat'] = pd.qcut(alvo,q=4 ).astype(str)
    
    dt['idade_cat'] = pd.qcut(dt['idade'], q=4).astype(str)
    
    
    # 3. Criação de novos dados.

    dt['tempo_emprego_cat'] = pd.qcut(dt['tempo_emprego'], q=4).astype(str)
    
    
    dt['renda_residencia'] = dt['tipo_renda'] + dt['tipo_residencia']
    
    dt['sexo_renda'] = dt['sexo'] + dt['tipo_renda']
    
    dt['imovel_renda'] = dt['posse_de_imovel'].astype(str) + dt['tipo_renda']
    
    
    atributos = ['idade_cat','tempo_emprego_cat','renda_residencia','sexo_renda','imovel_renda']
    
    # 4. Gráfico
    
    figura = plt.figure(figsize=(15,30))
    figura.suptitle('Análise das variáveis em relação à renda', fontsize=29)
    figura.subplots_adjust(top=0.95, # Espaço entre o título e os gráficos
                            hspace = 0.5)
    tamanho = (5,1)
    
        
    for i in range(tamanho[0]):
        pos = (i,0)
        ax = plt.subplot2grid(tamanho, pos)
    
        sns.countplot(data=dt, x=atributos[i],hue='renda_cat',ax =ax)
    
    return plt.show()


renda_analise_gerado(df, y_alvo)
```

```python
def make_it_better(data, alvo) -> None:
    """
    ----- Entrada ------
    data -> DataFrame contendo os atributos independentes
    alvo -> Series contendo o atributo dependente.
    
    ----- Objetivo -------
    O objetivo dessa função é misturar elementos do exercício anterior (Modificação de dados com base no p-value)
    como usar os novos atributos.
    
    ----- Funcionamento -------
    1. Seleção de atributos
        Com base no exercício anterior
        
    2. Binning
        Cortar um dado numérico em quartís, garantindo a proporção de 4 pedaços iguais, isso irá ajudar na análise 
        de distribuição 
        
    3. Criação de novos dados.      
    
    4. Gráfico.
    """  
    
    # 1. Configuração dos dados
    atributos = 'idade tempo_emprego qtd_filhos qt_pessoas_residencia sexo tipo_residencia posse_de_imovel tipo_renda educacao'
    atributos = atributos.split(' ')
    
    alvo = np.log(alvo)
    
    dt = data.copy()
    dt = dt.loc[:,atributos]
    
    dt['idade_cat'] = pd.qcut(dt['idade'], q=4).astype(str)

    dt['tempo_emprego_cat'] = pd.qcut(dt['tempo_emprego'], q=4).astype(str)
    
    dt['renda_residencia'] = dt['tipo_renda'] + dt['tipo_residencia']
    
    dt['sexo_renda'] = dt['sexo'] + dt['tipo_renda']
    
    dt['imovel_renda'] = dt['posse_de_imovel'].astype(str) + dt['tipo_renda']    
    
    atributos.extend(['idade_cat','tempo_emprego_cat','renda_residencia','sexo_renda','imovel_renda']) # Add novos elementos a lsita
    
    # Para o tipo de renda
    dt.loc[(dt['tipo_renda'] != 'Empresário') & (dt['tipo_renda'] != 'Pensionista') & (dt['tipo_renda'] != 'Servidor público'),'tipo_renda'] = 'outros'

    # Para o nível de educação
    dt.loc[(dt['educacao'] != 'Pós graduação') & (dt['educacao'] != 'Superior completo'), 'educacao'] = 'outros'
    
    # Para a quantidade de pessoas na residência diferente de 9 
    dt.loc[dt['qt_pessoas_residencia'].isin([6,7,9,15]), 'qt_pessoas_residencia'] = 6
    
    # Passar para o LabelEncoder
    for alterar in dt.select_dtypes(exclude=np.number):
        dt[alterar] = LabelEncoder().fit_transform(dt[alterar])
        
    # Divisão dos dados e criação do modelo.
    df_train, df_test, y_train, y_test = train_test_split(dt, alvo, test_size=0.25, random_state=100)
    modelo = Lasso(alpha=0).fit(df_train, y_train) # Treino
    predict = modelo.predict(df_test) # Previsão 
    r2_prev = r2_score(y_test, predict) # Resultado
    
    # Atributos modelo OLS
    numericos = 'idade + tempo_emprego + C(qtd_filhos) + C(qt_pessoas_residencia)'
    categoricos = '+ C(sexo) + C(posse_de_imovel) +  C(tipo_renda) + C(educacao) + C(idade_cat) + C(tempo_emprego_cat)'
    categoricos += '+ C(renda_residencia) + C(sexo_renda) + C(imovel_renda)'
    atributos = f'renda ~ {numericos} + {categoricos}'
         
    df_train['renda'] = y_train

    y, X = patsy.dmatrices(atributos, df_train)

    # Treinamento e análise de resultado
    model = sm.OLS(y, X).fit_regularized(method='elastic_net', refit=True, L1_wt = 1, alpha = 0)
    resultado = pd.DataFrame(model.summary().tables[0])
    resultado[3] = resultado[3].astype(str)
    aic = float(resultado.iloc[5,3])
    r2_sm = float(resultado.iloc[0,3].strip())
    adj_r2 = float(resultado.iloc[1,3].strip()) 
                      
    resumo = {'aic':aic, 'r2_sm':r2_sm, 'adj_r2':adj_r2, 'r2_prev':r2_prev}
    return resumo, model


resultado, modelo = make_it_better(df, y_alvo)
```

# Modelo de Árvore de decisão

```python
def modelo_arvore(data, alvo) -> None:
    """
    ----- Entrada ------
    data -> DataFrame contendo os atributos independentes
    alvo -> Series contendo o atributo dependente.
    
    ----- Objetivo -------
    O objetivo dessa função é criar 3 modelos baseado em árvores de decisão para obter o resultado de previsão do r2
    
    ----- Funcionamento -------
    1. Seleção de atributos
        Com base no exercício anterior
        
    2. Tratamento de dados 
    
    3. Treino e teste dos modelos de árvore de decisão criados.
    """  
    
    # 1. Configuração dos dados
    atributos = 'idade tempo_emprego qtd_filhos qt_pessoas_residencia sexo tipo_residencia posse_de_imovel tipo_renda educacao'
    atributos = atributos.split(' ')
    
    alvo = np.log(alvo)
    
    dt = data.copy()
    dt = dt.loc[:,atributos]
    
    dt['idade_cat'] = pd.qcut(dt['idade'], q=4).astype(str)

    dt['tempo_emprego_cat'] = pd.qcut(dt['tempo_emprego'], q=4).astype(str)
    
    dt['renda_residencia'] = dt['tipo_renda'] + dt['tipo_residencia']
    
    dt['sexo_renda'] = dt['sexo'] + dt['tipo_renda']
    
    dt['imovel_renda'] = dt['posse_de_imovel'].astype(str) + dt['tipo_renda']    
    
    atributos.extend(['idade_cat','tempo_emprego_cat','renda_residencia','sexo_renda','imovel_renda']) # Add novos elementos a lsita
    
    # Para o tipo de renda
    dt.loc[(dt['tipo_renda'] != 'Empresário') & (dt['tipo_renda'] != 'Pensionista') & (dt['tipo_renda'] != 'Servidor público'),'tipo_renda'] = 'outros'

    # Para o nível de educação
    dt.loc[(dt['educacao'] != 'Pós graduação') & (dt['educacao'] != 'Superior completo'), 'educacao'] = 'outros'
    
    # Para a quantidade de pessoas na residência diferente de 9 
    dt.loc[dt['qt_pessoas_residencia'].isin([6,7,9,15]), 'qt_pessoas_residencia'] = 6

    arvores = dict()
    
    # Passar para o LabelEncoder
    for alterar in dt.select_dtypes(exclude=np.number):
        dt[alterar] = LabelEncoder().fit_transform(dt[alterar])
        
    # Divisão dos dados e criação dso modelos.
    df_train, df_test, y_train, y_test = train_test_split(dt, alvo, test_size=0.25, random_state=100)
    
    # Árvore de decisão
    modelo = DecisionTreeRegressor().fit(df_train, y_train) # Treino
    predict = modelo.predict(df_test) # Previsão 
    r2_prev = r2_score(y_test, predict) # Resultado
    arvores['arvore_decisão'] = r2_prev
    
    # Floresta de decisão
    modelo = RandomForestRegressor(n_estimators=200 ,max_depth=None, min_samples_leaf=5).fit(df_train, y_train) # Treino
    predict = modelo.predict(df_test) # Previsão 
    r2_prev = r2_score(y_test, predict) # Resultado
    arvores['floresta_decisão'] = r2_prev
    
    # Gradiente de floresta de decisão
    modelo = GradientBoostingRegressor(n_estimators=150 ,max_depth=8, min_samples_leaf=2, learning_rate=0.09).fit(df_train, y_train) # Treino
    predict = modelo.predict(df_test) # Previsão 
    r2_prev = r2_score(y_test, predict) # Resultado
    arvores['floresta_gradiente'] = r2_prev
    
    return arvores

arvore = modelo_arvore(df, y_alvo)
```
