## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms (j na base)
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
df.info()
````

````python
def decomposicao(dt,coluna_data):
    """Funo para decomposio dos dados temporais."""
    # Cpia de segurana
    dados = dt.copy()
    # Decomposio dos dados temporais
    dados['Ms'] = dados[coluna_data].dt.month
    dados['Dia da semana'] = dados[coluna_data].dt.day_name()
    dados['Dia do ms'] = dados[coluna_data].dt.day
    dados['Final de Semana'] = dados[coluna_data].dt.dayofweek // 5 
    dados['Trimestre'] = dados[coluna_data].dt.quarter
    # Categorizao do valor NF.
    dados['Bin 5 Valo NF'] = pd.qcut(df['Valor NF'], q=5).astype(str)
    dados['Bin 10 Valo NF'] = pd.qcut(df['Valor NF'], q=10).astype(str)
    dados['Bin 20 Valo NF'] = pd.qcut(df['Valor NF'], q=20).astype(str)
    dados['Bin 50 Valo NF'] = pd.qcut(df['Valor NF'], q=50).astype(str)
    return dados
````

````python
dado = decomposicao(df,'Data Emisso')
dado.head()
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')
    
    rotulo_evento = tab.column
````

````python
class IV:
    def __init__(self):
        self.dt = dado.copy()
        self.iv = self.IV_total()
        
    def IV_calculo(self, coluna):
        # Criar uma tabela com o total de eventos (0) e no eventos (1).
        tab = pd.crosstab(self.dt[coluna], self.dt['flag_credito'], margins=True, margins_name='total')

        # Defino qual  a coluna de evento e no evento (somente o nome da coluna)
        rotulo_evento = tab.columns[0]
        rotulo_nao_evento = tab.columns[1]

        # Encontro a porcentagem de cada evento e no evento os divindindo pelos totais
        tab['pct_evento'] = tab[rotulo_evento]/tab.loc['total', rotulo_evento]
        tab['pct_nao_evento'] = tab[rotulo_nao_evento]/tab.loc['total', rotulo_nao_evento]

        # WOE
        tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
        # IV
        tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento) * tab.woe

        return tab['iv_parcial'].sum()
        
    def IV_total(self):
        colunas = dado.nunique()[dado.nunique() <= 50].index.tolist()
        colunas.remove('flag_credito')
        chave = list()
        for col in colunas:
            iv  = self.IV_calculo(col)
            chave.append([col, iv])


        return pd.DataFrame(data=chave, columns = ['Variavel','IV']).sort_values(by='IV',ascending=False)
    def classes(self,x):
        if x <0.02:
            return 'intil'
        elif (x>=0.02) & (x<0.1):
            return 'Fraco'
        elif (x>=0.1) & (x<0.3):
            return "Mdio"
        elif (x>=0.3) & (x<0.5):
            return "Forte"
    
    def grafico(self):
        resultado = self.iv
        resultado['Poder preditivo'] = resultado['IV'].apply(lambda x: self.classes(x))
        
        sns.set(style='whitegrid')
        plt.figure(figsize=(10,6))
        plt.suptitle('Poder Preditivo dos Atributos Categricos', fontsize=25)
        sns.barplot(data = resultado, orient='h', x='IV', y='Variavel', hue='Poder preditivo')
    
        return plt.show();
    
````

````python
IV().grafico()
````

## Descritiva simples no tempo

Para compor a anlise que vir a seguir, faa uma anlise descritiva bsica no tempo - utilize trimestre como agregao (ou uma mdia mvel por trimestre, como preferir) e avalie graficamente:

- A quantidade de notas registradas na base por perodo
- A quantidade de notas por categoria e por perodo
- A proporo de notas com crdito por categoria por perodo

````python
class analise:
    def __init__(self):
        self.g1 = dado.groupby(['Trimestre'])[['Trimestre']].count()
        self.g2 = dado.groupby(['categoria','Trimestre'])[['Trimestre']].count().unstack()    
        self.g2 = self.g2['Trimestre'].reset_index().melt(id_vars='categoria', value_vars=[1,2,3,4])
        
    def analise_0(self):
        plt.figure(figsize=(12,20))
        tamanho = (2, 1)

        # Grfico 1
        ax = plt.subplot2grid(
                tamanho,
                (0,0)
            )
        sns.barplot(x = self.g1.index, y=self.g1['Trimestre'], ax=ax)
        plt.title("Quantidade de notas registradas na base por perodo", fontsize=25)

        # Grfico 2
        ax = plt.subplot2grid(
                tamanho,
                (1,0)
            )

        sns.barplot(data=self.g2,x='value',hue='Trimestre',y='categoria',orient='h', ax=ax)   
        plt.title("Quantidade de notas por categoria e por perodo", fontsize=25)
        return plt.show()
    
    def analise_1(self):
        plt.figure(figsize=(12,30))
        tamanho = (5, 2)
        a1 = pd.get_dummies(dado, columns=['flag_credito'])
        a1 = a1.groupby(['categoria','Trimestre'])[['flag_credito_0','flag_credito_1']].sum().unstack()  
        a1.loc['total',:] = a1.sum(0)
        for col in a1.columns:
            a1[col] = a1.loc[:,col] / a1.loc['total',col]
        a1.reset_index().columns.tolist()
        a1 = a1.reset_index().melt(id_vars=[('categoria', '')], value_vars=a1.reset_index().columns.tolist()[1:])
        a1.columns = ['categoria','flag','trimestre','value']
        categorias = a1['categoria'].unique()
        plt.subplots_adjust(top=0.94, # Distncia do ttulo para os grficos.
                    hspace = 0.3, wspace=0.2)
        count = 0 
        plt.suptitle('Proporo de notas com crdito por categoria por perodo', fontsize=25)
        for row in range(5):
            for col in range(2):
                
                ax = plt.subplot2grid(
                    tamanho,
                    (row,col)
                )
                text = categorias[count]
                
                a2 = a1.loc[a1['categoria'] == categorias[count]]
                fig = sns.lineplot(data=a2, y='value',x='trimestre',hue='flag')
                
                plt.title(text, fontsize=15)
                #ax.set_xticklabels([1.,2.,3.,4.])
                count += 1       
                
        
````

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
def IV_calculo():
    # Criar uma tabela com o total de eventos (0) e no eventos (1).
    tab = pd.crosstab(dado['categoria'], dado['flag_credito'], margins=True, margins_name='total')

    # Defino qual  a coluna de evento e no evento (somente o nome da coluna)
    rotulo_evento = tab.columns[0]
    rotulo_nao_evento = tab.columns[1]

    # Encontro a porcentagem de cada evento e no evento os divindindo pelos totais
    tab['pct_evento'] = tab[rotulo_evento]/tab.loc['total', rotulo_evento]
    tab['pct_nao_evento'] = tab[rotulo_nao_evento]/tab.loc['total', rotulo_nao_evento]

    # WOE
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    # IV
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento) * tab.woe
    return tab.sort_values(by='woe', ascending=False)

IV_calculo()
````
