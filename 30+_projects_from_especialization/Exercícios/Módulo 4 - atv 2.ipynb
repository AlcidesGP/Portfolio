## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# Criando um dataframe com o total de questões

df_parametros = pd.DataFrame({'matematica':[24],'Português':[18],
                              'Geografia':[8],'inglês':[8],
                              'história':[8],'física':[12],
                              'Química':[12]})

display(df_parametros) 
````

````python
# Juntando os dados em um único dataframe

df = pd.concat([df_mat,df_por,df_geo,df_ing,df_his,df_fis,df_qui],axis=1) # Junção
df.columns = ['Mat_acertos','Por_acertos','Geo_acertos','Ing_acertos','His_acertos','Fis_acertos','Qui_acertos'] # Nome colunas

for col_p, col_dados in zip(df_parametros, df): # Loop usando o apply para cálcular a % de acerto.
    
    nome_coluna = f'porcentagem_acerto_{col_p}'
    df[nome_coluna] = df[col_dados].apply(lambda x: (x / df_parametros[col_p])*100).round(2)

df['Acertos_totais'] = df.iloc[:,:7].sum(1) # Somando cada linha de acertos
df_parametros['Total_questoes'] = df_parametros.iloc[:, :-1].sum(1) # Somando cada linha do total de questões.

df['Perc_acertos_totais'] = df['Acertos_totais'].apply(lambda x: (x / df_parametros['Total_questoes'])*100).round(2)

display(df.head())
````

````python
# 1) 
print(f"\n{' '*45} Percentual de acerto por matéria para cada aluno\n")
display(df.iloc[:,7:14])
````

````python
# 2) 
print(f"\n{' '*10} Total de acertos de cada aluno")
display(df[['Acertos_totais']])
````

````python
# 3) 

# Média da porcentagem de acerto
print(f"\n{' '*10}Média da Porcentagem de acertos")
display(df.iloc[:,7:14].describe().T[['mean']].round(2))

# Média de acertos
print(f"\n{' '*15}Média de acertos")
df.iloc[:,:7].describe().T[['mean']]
````

````python
# 4) 
total = df.loc[df['Acertos_totais'] > 45].shape[0]
print(f"\n{' '*10} Um total de {total} alunos tiraram nota acima de 45 e passaram.\n")
````
