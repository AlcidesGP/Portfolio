# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

<a id=1></a>
1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
media = 15
std = 0.1
limite_sup = 15.2
prob = stats.norm.cdf(limite_sup,loc=media, scale=std)
print(f"A probabilidade de um pacote selecionado de forma aleatria tenha mais de 15,2 kgs  de {(1-prob)*100:.2f} %")
```

<a id=2></a>
2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
media = 15
std = 0.1
porcentagem = .95
valor = stats.norm.ppf(porcentagem, loc=media, scale=std)
print(f"95 % das raes vo ter um peso menor que {valor:.2f} Kgs")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

4. Faa uma visualizao grfica dos dados do item anterior.

```python
Y=np.linspace(-6, 6, 200)
# Distribuio Normal
media = 0
var = 1
prob_normal = stats.norm.cdf(Y, loc=media, scale=var)

# Distribuio T-studdent
grau_liberdade = 5
prob_t = stats.t.cdf(Y, df=grau_liberdade)

# Grfico de comparao entre a probabilidade para cada valor
plt.figure(figsize=(10,5))
sns.scatterplot(x=prob_normal,y=Y, label='Normal')
sns.scatterplot(x=prob_t,y=Y, label='T-Student')
plt.title('Comparao de Distribuio Normal e T-Student')
plt.xlabel('Probabilidade')
plt.ylabel('Valores de Y');
```

5. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Distribuio Normal
media = 0
var = 1
percentil_norm = stats.norm.ppf(Y, loc=media, scale = var)

# Distribuio T
grau_liberdade = 5
percentil_t = stats.t.ppf(Y, df=grau_liberdade)
```

<a id=6></a>

6. Faa um grfico com os dados do item anterior.

```python
plt.figure(figsize=(10,5))
sns.kdeplot(percentil_norm, label='normal')
sns.kdeplot(percentil_t, label='T-student')

plt.legend()
plt.xlabel('Valores')
plt.ylabel('Densidade Kernel Estimada')
plt.title('Estimativa de Densidade Kernel (KDE)');
```
