# Main 

```python
ocr = PaddleOCR(lang='pt')
```

```python
# Caminho para a imagem
image_path = './livro.jpg'

# Extrair o texto da imagem
result = ocr.ocr(image_path)

# Imprimir o texto extraído
for line in result:
    for word in line:
        print(word[1][0])
```
