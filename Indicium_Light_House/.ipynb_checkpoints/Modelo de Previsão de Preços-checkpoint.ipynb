```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
import math
from tqdm import tqdm
from sklearn.base import BaseEstimator, TransformerMixin
import holidays
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
import statsmodels.formula.api as smf
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import ElasticNet, HuberRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
```

```python
class ImportarDados():
    @property
    def base_original(self):
        dados = pd.read_csv("./Dados/teste_indicium_precificacao.csv")
        eliminar_zeros = dados.loc[dados['price'] == 0].index
        dados = dados.drop(axis=1, index=eliminar_zeros)
        return dados
    
    @property
    def base_universidades(self):
        dados = pd.read_csv("./Dados/city_uni.csv")
        dados.columns = [i.lower() for i in dados.columns]
        return dados

    def base_subway(self) -> pd.DataFrame:
        """
        Agora podemos trabalhar com as outras regiões de New York, sendo todas essas interligadas por redes de metrô. 
        Para essa situação a abordagem será diferente, pois o governo america disponibiliza no site as informações sobre 
        a latitude e longitude de cada entrada de estação que pode ser coletada através de uma API,
        então usaremos esses dados para poder calcular qual é a entrada de metrô mais próxima de cada hostel.
        """
        # URL do site do governo de Nova York com os dados das entradas das estações
        tabela = pd.read_csv('https://data.ny.gov/resource/i9wp-a4ja.csv')

        #Os dados de interesse são somente a região e as coordenadas da entrada. Então já podemos fazer o filtro.
        tabela = tabela.loc[:,['station_name','borough','entrance_latitude','entrance_longitude']]

        # O dados da região estão no formato reduzido, então a função map será usada para ajustar.
        tabela['borough'] = tabela['borough'].map(
                                    {
                                        'M':'Manhattan',
                                        'Q':'Queens',
                                        'B':'Brooklyn',
                                        'Bx':'Bronx'
                                    }
                                )

        # Abaixo estou removendo os dados duplicados e resetando o índice.
        tabela = tabela.drop_duplicates().reset_index(drop=True)
        tabela.columns = ['station_name','borough','latitude','longitude']

        return tabela
    
    def obter_geoloc(self, station: str) -> list:
        """ 
        O objetivo dessa função é coletar a posição geográfica de cada estação de trem da Staten Island. 
        Staten Island é um distrito mais afastado de Nova York e só apresenta linhas de trem, sem nenhum metrô. 
        """
        geolocator = Nominatim(user_agent="localizador") # Função que permite retornar a latitude e longitude de um local somente com o texto.
        loc = geolocator.geocode(f'{station}, staten island ,New York', language='en') # Aplicando a função
        
        
        return [station, loc.latitude, loc.longitude] # retornando as informações.
    
    
    def base_trem(self) -> pd.DataFrame:
        # Importando tabela das estações de trem de staten island
        staten_island = pd.read_html('https://en.wikipedia.org/wiki/List_of_Staten_Island_Railway_stations', header=0)[1]

        # Separando somente o nome de cada estação
        staten_island = staten_island['Name']

        # Criar lista para salvar a estação, latitude e longitude.
        staten_resultados = list()
        for col in staten_island:
            staten_resultados.append(self.obter_geoloc(col))

        # Dataset contendo a latitude e longitude.   
        df_staten = pd.DataFrame(staten_resultados, columns=['station_name','latitude','longitude'])
        df_staten['borough'] = ['Staten Island' for i in range(df_staten.shape[0])]
        return df_staten
    
    @property
    def base_stations(self) -> pd.DataFrame:
        # Concatenando ambas os datasets.
        df_stations = pd.concat([self.base_trem(), self.base_subway()], ignore_index=True)
        return df_stations
    
    
    def obter_geoloc_turismo(self, local: str) -> list:
        """ O objetivo dessa função é coletar a posição geográfica de 12 pontos turísticosda cidade de New York"""

        # A funçaõ abaixo vai permitir converter endereços em uma localização geográfica.
        geolocator = Nominatim(user_agent="localizador")
        loc = geolocator.geocode(f'{local}, New York', language='en')

        return [local, loc.latitude, loc.longitude]
    
        # Principais 12 pontos turísticos.
  
    
    @property
    def base_turismo(self):
        turismo = ['Empire State Building','Grand Central Terminal','Memorial','The Metropolitan Museum',
                   'Statue of Liberty','Times Square','Broadway', 'Wall Street','One World Observatory',
                   'Brooklyn Museum','Central Park','Museum of Natural History']
        # Criar uma lista para salvar os resultados. Rodar um looping para cada localidade.
        locais = list()
        for i in turismo:
            locais.append(self.obter_geoloc_turismo(i))
        df_turismo = pd.DataFrame(locais, columns=['turism_place','latitude','longitude'])
        return df_turismo
```

```python
class ColetaDadosDistancia(BaseEstimator, TransformerMixin):   
    
    def metodo_euclidiano(self, subway,coluna, df_referencial):
        menor_distancia = float('inf')
        ponto_mais_proximo = None
        dt_subway = subway.copy()
        if coluna == 'borough':
            dt_subway = dt_subway.loc[dt_subway[coluna] == df_referencial[0]]
            
        ponto_referencia = df_referencial[1:]
        pontos = dt_subway[['latitude','longitude']].to_numpy()
        for ponto in pontos:
            distancia = self.calcular_distancia(ponto, ponto_referencia)
            if distancia < menor_distancia:
                menor_distancia = distancia
                ponto_mais_proximo = ponto

        
        return geodesic((ponto_mais_proximo[0],ponto_mais_proximo[1]), (ponto_referencia[0], ponto_referencia[1])).km
    
    def calcular_distancia(self, ponto1, ponto2):
        x1, y1 = ponto1
        x2, y2 = ponto2
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, dados):
        
        dt = dados.copy()
        station = ImportarDados().base_stations
        uni = ImportarDados().base_turismo
        turismo = ImportarDados().base_turismo
        
        # ==== Coletando dados de metro
        
        resultado = list()
        for n,valores in tqdm(dt.iterrows(), total=len(dt), desc="Distancia Metrô"):
            resultado.append(ColetaDadosDistancia().metodo_euclidiano(station,'borough' ,[valores['bairro_group'], valores['latitude'], valores['longitude']]))
        dt['distacia_metro'] = resultado
        
        # ==== Coletando dados Universidade
        
        resultado_uni = list()
        for n,valores in tqdm(dt.iterrows(), total=len(dt), desc="Distancia Universidade"):
            resultado_uni.append(ColetaDadosDistancia().metodo_euclidiano(uni,'city' ,[valores['bairro_group'], valores['latitude'], valores['longitude']]))
        dt['distacia_uni'] = resultado_uni
        
        # ==== Coletando dados Turismo
        for i, ponto_turistico in tqdm(turismo.iterrows(), total=len(turismo), desc="Distância Pontos Turísticos"):
            resultado_turismo = list()
            for n,valores in dt.iterrows(): 
                resultado_turismo.append(geodesic((ponto_turistico['latitude'],ponto_turistico['longitude']), (valores['latitude'], valores['longitude'])).km)
            nome = ponto_turistico['turism_place'].lower().replace(' ','_')
            dt[nome] = resultado_turismo
        
        return dt
        
        
```

```python
class TransformacaoTemporal(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        ## Passar para o formato Date Time.
        date_time = pd.to_datetime(X['ultima_review']).copy()     
        df_temporal = X.copy()
        
        ## Extração de dados temporais
            # Dia
        df_temporal['dia'] = date_time.dt.day
            # Mês
        df_temporal['mes'] = date_time.dt.month
            # Dia da semana
        df_temporal['dia_semana']  = date_time.dt.day_name()
            # Final de semana ou não.
        df_temporal['fim_de_semana']  = date_time.dt.day_of_week.apply(lambda x: 'final_de_semana' if x in [5, 6] else ('dia_de_semana' if pd.notna(x) else 'Não informado'))
            # Se é feriado ou não.
        country_code = 'US'
        feriados = holidays.CountryHoliday(country_code)
        # Função abaixo irá retornar se é feriado, se não é ou se não foi registrado.
        df_temporal['feriado']  = date_time.apply(lambda x: 'Feriado' if x in feriados else('Não Feriado' if pd.notna(x) else 'Não informado'))
        
        # Preencher vazios por falta de review nos numéricos
        for col in ['ano','dia','mes','reviews_por_mes']:
            df_temporal[col] = df_temporal[col].fillna(0)

        # Preencher vazios por falta de review no nominal.   
        df_temporal['dia_semana'] = df_temporal['dia_semana'].fillna('Não informado')
        
        df_temporal = df_temporal.drop(axis=0,columns=['ultima_review'])
        
        return df_temporal
```

```python
class AjustesDados(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        # Cópia de segurança
        dt = X.copy()
        
        # ========== Remover dados
        dt = dt.drop(axis=0, columns=['id','nome','host_name'])
        
        if 'price' in dt.columns:
            remover = dt.loc[dt['price'] == 0].index
            dt = dt.drop(axis=1, index=remover)
        dt = dt.rename(columns={'bairro_group': 'distrito'})
        return dt
    
```

```python
class AvaliacaoDados():
    def hl_modelo(self, sumario):

            tabela = pd.DataFrame(sumario.tables[1], columns = ['Atributos','coef','std','t','P>|t|','[0.025','0.975'])
            tabela.drop(0,axis=0, inplace=True)        
            for i in tabela:
                tabela[i] = tabela[i].astype(str).astype(str)
           # tabela.set_index('Atributos',inplace=True)
            tabela.set_index('Atributos', inplace=True)
            for i in tabela:
                tabela[i] = tabela[i].astype(str).astype(float)

            #display(tabela.style.highlight_between(left=0, right=0.1,subset='P>|t|'))
            return tabela
    def matrix_corr(self, entrada):
        plt.figure(figsize=(15,15))
        correl = abs(entrada.corr(method='spearman'))
        correl[(correl <.8) & (correl > -.8)] = 0
        col = correl.sum(1)[correl.sum(1)>1].index
        correl = correl.loc[col,col]
        mask = np.triu(np.ones_like(correl, dtype=bool))
        sns.heatmap(correl[col], mask=mask, cbar=False, annot=True)
```

O que buscamos ao analisar o P>|t|, basicamente ele nos retornará se aquela variável teve uma significância da dispersão da variância que contribuiu para a predição do modelo, permitindo assim realizarmos um feature selection em cima dos dados. Então vamos averiguar quais variáveis ficaram com um valor inferior a 0.1 (geralmente usa-se 0.05, mas como quero dar uma chance a algumas variáveis, vou subir um pouco a régua)

Agora com a definição do R2 base a ser superado e quais tratamentos as variáveis irão receber, podemos então refazer a função de tratamento dos dados e passar um novo pipeline.

```python
class FeatureEngineering():
    def transform(self, X, y=None):
        colunas_remover = ['one_world_observatory','statue_of_liberty','museum_of_natural_history','the_metropolitan_museum']
        bairros_alterar = ['Arden Heights', 'Arverne',  'Bath Beach',  'Battery Park City',  'Bay Terrace, Staten Island',  'Baychester',
 'Bedford-Stuyvesant',  'Belle Harbor', 'Belmont', 'Bensonhurst', 'Bergen Beach', 'Borough Park', 'Brighton Beach', 'Brooklyn Heights',
 'Brownsville', "Bull's Head", 'Bushwick', 'Canarsie', 'Castle Hill', 'Castleton Corners', 'Chelsea', 'Claremont Village', 'Clason Point',
 'Co-op City',  'College Point', 'Columbia St', 'Coney Island', 'Cypress Hills', 'Downtown Brooklyn', 'Dyker Heights', 'East Flatbush',
 'East Morrisania', 'East New York', 'East Village', 'Edenwald', 'Eltingville', 'Emerson Hill', 'Far Rockaway', 'Fieldston',
 'Flatbush', 'Fordham', 'Gramercy', 'Graniteville', 'Gravesend', 'Great Kills', 'Greenpoint', 'Greenwich Village', 'Grymes Hill', 'Howard Beach', 'Howland Hook',
 'Huguenot', 'Inwood', 'Kensington', 'Kew Gardens', 'Kingsbridge', 'Laurelton', 'Lighthouse Hill', 'Longwood', 'Manhattan Beach', 'Marble Hill', 'Mariners Harbor',
 'Middle Village', 'Midwood', 'Mill Basin', 'Morris Park', 'Morrisania', 'Navy Yard', 'Neponsit', 'New Brighton', 'New Dorp',
 'New Springville', 'Nolita', 'North Riverdale', 'Norwood', 'Oakwood', 'Olinville', 'Parkchester', 'Pelham Bay', 'Pelham Gardens',
 'Port Richmond', 'Red Hook', 'Rego Park', 'Richmond Hill', 'Richmondtown', 'Rosedale', 'Rossville', 'Schuylerville', 'Sheepshead Bay', 'Shore Acres',
 'Silver Lake', 'South Beach', 'South Ozone Park', 'Spuyten Duyvil', 'Stuyvesant Town', 'Sunset Park', 'Throgs Neck', 'Todt Hill',
 'Unionport', 'University Heights', 'Van Nest', 'Wakefield', 'Washington Heights', 'West Farms', 'West Village', 'Westchester Square', 'Westerleigh', 'Whitestone',
 'Williamsbridge', 'Willowbrook', 'Woodlawn']
        dias_alterar = ['Monday', 'Thursday', 'Tuesday', 'Wednesday']
        dt = X.copy()
        dt.drop(axis=0, columns=colunas_remover)
        dt['bairro'] = dt['bairro'].map(lambda x: x if x not in bairros_alterar else 'outros')
        dt['dia_semana'] = dt['dia_semana'].map(lambda x: x if x not in dias_alterar else 'outros')
        dt['distrito'] = dt['distrito'].map(lambda x: x if x not in ['Manhattan','Brooklyn'] else 'Manhattan_Brooklyn')
        
        remover = ['one_world_observatory','statue_of_liberty','museum_of_natural_history','the_metropolitan_museum']
        dt = dt.drop(axis=0, columns=remover)
        id = dt.pop('host_id')
        
        transformar = ['latitude','longitude','minimo_noites','numero_de_reviews','reviews_por_mes',
                       'calculado_host_listings_count','disponibilidade_365','distancia_metro',
                       'distacia_uni', 'empire_state_building', 'grand_central_terminal', 'memorial',
                        'times_square','broadway', 'wall_street',  'brooklyn_museum','central_park']
        
        dt.loc[:,transformar] = StandardScaler().fit_transform(dt[transformar])
        
        transformador = PowerTransformer(method='yeo-johnson')
        for col in transformar:
            dt[col] = transformador.fit_transform(dt[col].values.reshape(-1, 1))
            
        
        for col in dt.select_dtypes(exclude=np.number):
            dt[col] = LabelEncoder().fit_transform(dt[col])
        if y is not None:
            return dt, y
        else:
            return dt
        
```

```python
class ModelosPrevisao:
    def __init__(self, X_train, X_test, y_train,  y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
        self.elastic_net = ElasticNet()
        self.huber_regressor = HuberRegressor()
        self.knn_regressor = KNeighborsRegressor()
        self.random_forest = RandomForestRegressor()
        self.gradient_boost = GradientBoostingRegressor()
        self.adaboost = AdaBoostRegressor()
        self.lightgbm = lgb.LGBMRegressor()
        
    def train_model(self):
        self.elastic_net.fit(self.X_train, self.y_train)
        self.huber_regressor.fit(self.X_train, self.y_train)
        self.knn_regressor.fit(self.X_train, self.y_train)
        self.random_forest.fit(self.X_train, self.y_train)
        self.gradient_boost.fit(self.X_train, self.y_train)
        self.adaboost.fit(self.X_train, self.y_train)
        self.lightgbm.fit(self.X_train, self.y_train)
        
        predictions = {
            'Elastic Net':self.elastic_net.predict(self.X_test),
            'Huber':self.elastic_net.predict(self.X_test),
            'Knn':self.elastic_net.predict(self.X_test),
            'Random Forest':self.elastic_net.predict(self.X_test),
            'Gradient Boost':self.elastic_net.predict(self.X_test),
            'Ada Boost':self.elastic_net.predict(self.X_test),
            'Light GBM':self.elastic_net.predict(self.X_test),
            
        }
        
        return predictions
    
    def metricas(self, y_true, y_pred):     
        r2 = r2_score(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        return r2, rmse
    
    def resultados(self):
        total = {}
        for nome, valores in self.train_model().items():
            r2, rmse = self.metricas(self.y_test, valores)
            total[nome] = [r2, rmse]
        return total
```

```python
class DivisaoDados:
    def dividir(self, dado_entrada, target):
        X = dado_entrada.copy()
        y = X.pop(target)
        X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=20, train_size=0.8)
        X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, random_state=20, train_size=0.95)
        print(f"Tamanho dados treino {X_train.shape}\nTamanho dados teste {X_test.shape}\nTamanho dados Validação {X_val.shape}")
        return X_train, X_test, X_val, y_train, y_test, y_val
    
    def tratamento_dados(self, X, y=None):
        if y is not None:
            X, y = FeatureEngineering().transform(X,y)
            return X, y
        else:
            X = FeatureEngineering().transform(X)
            return X,y
        
    def novos_dados(self, X):
        X = FeatureEngineering().transform(X)
        return X
```

```python
class ModelosML:
    def __init__(self, entrada):
        self.Xtr = entrada[0]
        self.Xte = entrada[1]
        self.ytr = entrada[2]
        self.yte = entrada[3]
    
    def metricas(self, y_pred):
        r2 = r2_score(self.yte, y_pred)
        rmse = np.sqrt(mean_squared_error(self.yte, y_pred))
        return [r2, np.exp(rmse)]
        
    def modelos(self):
        self.elastic_net = ElasticNet().fit(self.Xtr, self.ytr)
        self.huber_regressor = HuberRegressor(max_iter=1000).fit(self.Xtr, self.ytr)
        self.knn_regressor = KNeighborsRegressor().fit(self.Xtr, self.ytr).fit(self.Xtr, self.ytr)
        self.random_forest = RandomForestRegressor().fit(self.Xtr, self.ytr)
        self.gradient_boost = GradientBoostingRegressor().fit(self.Xtr, self.ytr)
        self.adaboost = AdaBoostRegressor().fit(self.Xtr, self.ytr)
        self.lightgbm = lgb.LGBMRegressor(verbosity=-1).fit(self.Xtr, self.ytr)
        
    def train_model(self):
        self.modelos()
        predictions = {
            'Elastic Net':self.metricas(self.elastic_net.predict(self.Xte)),
            'Huber':self.metricas(self.huber_regressor.predict(self.Xte)),
            'Knn':self.metricas(self.knn_regressor.predict(self.Xte)),
            'Random Forest':self.metricas(self.random_forest.predict(self.Xte)),
            'Gradient Boost':self.metricas(self.gradient_boost.predict(self.Xte)),
            'Ada Boost':self.metricas(self.adaboost.predict(self.Xte)),
            'Light GBM':self.metricas(self.lightgbm.predict(self.Xte))
            
        }
        return predictions
    
    def relatorio(self,texto):
        resultado = self.train_model()
        print(f"{' '*10}{texto}")
        for nome, item in resultado.items():
            print(f"Para o modelo {nome} o resultado de R2 = {item[0]*100:.2f}% e RMSE = {item[1]}")
```

```python
def simples(dados):
    dt = dados.copy()
    dt = dt.drop(axis=0, columns=['id','nome','host_id','host_name','ultima_review'])
    dt['reviews_por_mes'] = dt['reviews_por_mes'].fillna(0)
    
    dt = pd.get_dummies(dt, columns=dt.select_dtypes(exclude=np.number).columns, drop_first=True)
    
    #for col in dt.select_dtypes(exclude=np.number):
     #   dt[col] = LabelEncoder().fit_transform(dt[col])
    
    return dt
```

```python
df_simples = simples(df)
```

```python
X_train, X_test, X_val, y_train, y_test, y_val = DivisaoDados().dividir(df_simples,'price')
```

```python
ModelosML([X_train, X_test, y_train, y_test]).relatorio('Com o host_id')
```

```python
ModelosML([X_train, X_test, y_train, y_test]).relatorio('Com o nome_host')
```
