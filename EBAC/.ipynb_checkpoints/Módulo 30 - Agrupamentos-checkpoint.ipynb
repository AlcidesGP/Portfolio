# Métodos hierárquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

import scipy.cluster.hierarchy as shc
```

<a id=1></a>
## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variáveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variáveis para média zero e desvio padrão 1
4. Faça um agrupamento hierárquico dividindo essa base em 3 grupos
5. Faça o Dendrograma

```python
# Carregar o conjunto de dados dos pinguins
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Averguar Vazios
penguins.isnull().sum().sum()
```

```python
# Tratar vazios
df = penguins.dropna()
```

```python
# Verificar se há duplicados
df.duplicated().sum()
```

```python
def cluster(dt) -> None:
    # Copiando base de dados, selecionando quantitativos e padronizando-os.
    original = dt.copy()
    data = dt.copy()
    data = data.select_dtypes(include=np.number)
    data.iloc[:,:] = StandardScaler().fit_transform(data)
    
    # Criando modelo agloremativo.
    cluster = AgglomerativeClustering(n_clusters=3, linkage='complete')
    cluster.fit(data)
    data['grupos'] = cluster.labels_
    original['grupos'] = cluster.labels_
    
    # Criando Dendograma.
    shc.dendrogram(shc.linkage(data, method='complete'))
    
    
    # Retornando Grupos ao dado original 
    return original

analise = cluster(df)
```

## 2) Avalie os grupos

Descreva os grupos:

- Quantos pinguins de cada espécie há em cada grupo?
- Descreva as variáveis quantitativas de cada grupo com as técnicas que você preferir.

```python
def grup_aval():
    print(f"\n{' '*15}Grupos x Especies")
    display(analise.groupby(['grupos','species'])[['species']].count())
    print(f"\n{' '*10}Média das variáveis quantitativas\n")
    display(analise.groupby('grupos')[analise.select_dtypes(include=np.number).columns.tolist()].mean())
    
    
grup_aval()
```

## 3) Alterando o agrupamento

O Dendrogarma sugere que podemem haver grupos interessantes se solicitarmos mais que 3 grupos. Se quisermos testar o agrupamento com mais de 3 grupos, quantos grupos o Dendrograma te sugere? (Dica: não faça mais que 6 grupos).

- Faça o agrupamento hierárquico semelhante ao anterior, mas com o número de grupos que você escolheu.
- Analise esses agrupamentos
    - Quantos pinguins de cada espécie há em cada grupo?
    - Quantos pinguins de cada espécie machos e fêmeas há em cada grupo?

```python
def cluster(dt) -> None:
    # Copiando base de dados, selecionando quantitativos e padronizando-os.
    original = dt.copy()
    data = dt.copy()
    data = data.select_dtypes(include=np.number)
    data.iloc[:,:] = StandardScaler().fit_transform(data)
    
    # Criando modelo agloremativo.
    cluster = AgglomerativeClustering(n_clusters=5, linkage='complete')
    cluster.fit(data)
    data['grupos'] = cluster.labels_
    original['grupos'] = cluster.labels_
    
    # Criando Dendograma.
    shc.dendrogram(shc.linkage(data, method='complete'))
    print(f"\n{' '*15}Grupos x Especies")
    display(original.groupby(['grupos','species'])[['species']].count())
    print(f"\n{' '*10}Média das variáveis quantitativas\n")
    display(original.groupby('grupos')[original.select_dtypes(include=np.number).columns.tolist()].mean())
    
    # Retornando Grupos ao dado original 
    return original

analise = cluster(df)
```

# 4) Conclusões
- A divisão com 5 grupos fica mais percetível analisando a média dos dados quantitativos, aonde percebe-se os diferentes valores para cada grupo de dados analisado. Quando a analise busca avaliar as espécies de pinguins fica perceptível as classes que foram divididas, sendo a 1 e 3 dividindo a mesma espécie, no qual a média das características quantitativas são menores para a classe 3, ou seja, acaba sendo uma subdivisão da espécie. Algo semelhante ocorre nos grupos 2 e 4, com o último apresentando médias maiores que o grupo 2, mas ambos os dados ainda pegando outliers de outra espécie.
