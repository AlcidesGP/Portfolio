````python
def correlacao_quan() -> None:
    numericos = df.select_dtypes(include=np.number)
    # Fazendo uma correlação não linear.
    numericos_cor = numericos.corr(method='pearson')
    # Criação da variação de cor.
    figura = plt.figure(figsize=(15,8))
    cmap = sns.diverging_palette(h_neg=0, h_pos=240, as_cmap=True, sep = 1, n=14, center='light') # Ajuste de cor
    
    mask = np.triu(np.ones_like(numericos_cor, dtype=bool)) # Aarranjo da matriz para somente mostrar os valores uma vez;
    
    corre = sns.heatmap(numericos_cor, center = 0, cmap = cmap, annot=True, mask=mask) # Plot
    
    return plt.show()
    
    
correlacao_quan()
````

3) Na sua opinião, há outliers na variável renda?

````python
def caixa() -> None:
    figura = plt.figure(figsize=(15,7))
    figura.suptitle('Gráfico de caixa', fontsize=25)
    dados = df.copy()
    
    caixa = sns.boxplot(dados['renda'], orient='h', width=0.3)
    
    quantiles = dados['renda'].quantile([0.25, 0.75]).values.tolist() # Separação dos quantiles.
    LI = quantiles[0] - ((quantiles[1] - quantiles[0])*1.5) # Definir limite inferior
    LS = quantiles[1] + ((quantiles[1] - quantiles[0])*1.5) # Definir limte superior
    if LI < dados['renda'].min(): # Loop que vai reposicionar em caso de valores abaixo/acima do cálculado.
        LI = dados['renda'].min()
    if LS > dados['renda'].max():
        LS = dados['renda'].max()

    posse = ['right', 'left'] # Reajustar a posição no boxplot.
    for ix,poss in zip([int(LI),int(LS)],posse): # Plotar no gráfico.
        caixa.annotate(round(ix,2), xy=(ix, 0.13), fontsize=12, color='black',horizontalalignment=poss)
    for ix,poss in zip(quantiles,posse):
        caixa.annotate(round(ix,2), xy=(ix, -0.17), fontsize=12, color='black',horizontalalignment=poss)
    
    
    
    return plt.show()


caixa()
````

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
# Passar o log na variável renda e salvar como um novo atributo.
df['log_renda'] = np.log(df['renda'])
````

````python
def histograma() -> None:
    figura = plt.figure(figsize=(15,10))
    figura.suptitle('Comparação da dispersão da renda com log da renda.', fontsize=20)
    tamanho = (2,2)
    
    
    # Gráfico 1
    ax = plt.subplot2grid(tamanho, (0,0))
    sns.histplot(df['renda'])
    
    # Gráfico 2
    ax = plt.subplot2grid(tamanho, (0,1))
    sns.histplot(df['log_renda'])
    
    # Gráfico 3
    ax = plt.subplot2grid(tamanho, (1,0), colspan=2)
    
    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    ax.set_xticks([])
    ax.set_yticks([])
    texto = """A comparação acima demonstra que o log transformou os \ndados de uma distribuição Gamma para uma normal"""
    ax.annotate(texto, xy=(0.5,0.5), horizontalalignment='center',fontsize=20)
    
    return plt.show()

histograma()
````

````python
def correlacao() -> None:
    figura = plt.figure(figsize=(15,10))
    figura.suptitle('Gráfico de correlação', fontsize=20)
    tamanho = (2,1)
    
    # Gráfico 1
    ax = plt.subplot2grid(tamanho, (0,0))
    numericos = df.select_dtypes(include=np.number)
    # Fazendo uma correlação não linear.
    numericos_cor = numericos.corr(method='pearson')
    # Criação da variação de cor.
    figura = plt.figure(figsize=(15,8))
    cmap = sns.diverging_palette(h_neg=0, h_pos=240, as_cmap=True, sep = 1, n=14, center='light') # Ajuste de cor
    mask = np.triu(np.ones_like(numericos_cor, dtype=bool)) # Aarranjo da matriz para somente mostrar os valores uma vez;
    corre = sns.heatmap(numericos_cor, center = 0, cmap = cmap, annot=True, mask=mask, ax=ax) # Plot
    
    # Gráfico 2
    ax = plt.subplot2grid(tamanho, (1,0))
    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    ax.set_xticks([])
    ax.set_yticks([])
    texto = """- Analisando a tabela de correlação acima, observa-se que não houve uma mudança muito 
            significativa em relação a força e direção dos atributos com a renda"""
    ax.annotate(texto, xy=(0.5,0.5), horizontalalignment='center',fontsize=15)
    
    return plt.show()


correlacao()
````

````python
def caixa() -> None:
    figura =plt.figure(figsize=(15,10))
    figura.suptitle('Gráfico de caixa', fontsize=20)
    tamanho = (2,1)
    dados = df.copy()
    
    
    # Gráfico 1
    ax = plt.subplot2grid(tamanho, (0,0))
    
    caixa = sns.boxplot(dados['log_renda'], orient='h', width=0.3)
    
    quantiles = dados['log_renda'].quantile([0.25, 0.75]).values.tolist() # Separação dos quantiles.
    LI = quantiles[0] - ((quantiles[1] - quantiles[0])*1.5) # Definir limite inferior
    LS = quantiles[1] + ((quantiles[1] - quantiles[0])*1.5) # Definir limte superior
    if LI < dados['log_renda'].min(): # Loop que vai reposicionar em caso de valores abaixo/acima do cálculado.
        LI = dados['log_renda'].min()
        
    if LS > dados['log_renda'].max():
        LS = dados['log_renda'].max()
        
    posse = ['right', 'left'] # Reajustar a posição no boxplot.
    for ix,poss in zip([round(LI,2),round(LS,2)],posse): # Plotar no gráfico.
        caixa.annotate(round(ix,2), xy=(ix, 0.13), fontsize=12, color='black',horizontalalignment=poss)
    for ix,poss in zip(quantiles,posse):
        caixa.annotate(round(ix,2), xy=(ix, -0.17), fontsize=12, color='black',horizontalalignment=poss)
    
    
    # Gráfico 2
    ax = plt.subplot2grid(tamanho, (1,0))

    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    ax.set_xticks([])
    ax.set_yticks([])
    texto = """- Após a transformação de log percebe-se que os dados permaneceram com outliers, sendo os valores abaixo de 5.98
e acima de 10.24"""
    ax.annotate(texto, xy=(0.5,0.5), horizontalalignment='center',fontsize=15)    
    
    return plt.show()
    
    
caixa()
    
````
