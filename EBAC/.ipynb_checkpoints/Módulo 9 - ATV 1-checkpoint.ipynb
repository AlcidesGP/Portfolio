# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_raw.groupby('munResNome')[['IDADEMAE','IDADEPAI']].mean().sort_values(by='IDADEPAI', ascending=False)
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_raw.loc[sinasc_raw['DTNASC'].str.contains('10-27')].groupby(['SEXO','ESCMAE'])[['PESO']].mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
cidade = sinasc_raw.groupby('munResNome')[['ORIGEM']].count().sort_values(ascending=True, by='ORIGEM')
print(f'Cidade que nasceu menos pessoas foi {cidade.index[0]}, com um total de {cidade.values[0,0]}')

print(f'\nComo esse dado aparenta mais ser algum tipo de ERRO, o município levado em consideração será o segundo, no caso {cidade.index[1]}')
print(f'Que apresentou um total de {cidade.values[1,0]} nascimentos.\n')


sinasc_raw.loc[sinasc_raw['munResNome'] == cidade.index[1]].groupby(['munResNome'])[['IDADEPAI','IDADEMAE']].agg(['mean','max','min']).T
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
cidade = sinasc_raw.groupby('munResNome')[['ORIGEM']].count().sort_values(ascending=False, by='ORIGEM')
print(f'Cidade que nasceu menos pessoas foi {cidade.index[0]}, com um total de {cidade.values[0,0]}')


display(sinasc_raw.loc[sinasc_raw['munResNome'] == cidade.index[0]].groupby(['munResNome'])[['QTDFILVIVO']].agg(['mean','max','min']).T)
display(sinasc_raw.loc[sinasc_raw['munResNome'] == cidade.index[0]].groupby(['munResNome'])[['IDADEPAI']].agg(['mean','max','min']).T)
```
