#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
def analise(data: list, elemento: str, intervalo: int) -> None:
    dt = data.copy()
    dt_mm = dt[elemento].rolling(intervalo).mean().dropna()
    dt_std = dt[elemento].rolling(intervalo).std().dropna()
    
    dt_mm = dt_mm.stack().reset_index()
    dt_mm.columns = ['Data','Ativos','MM']
    dt_std = dt_std.stack().reset_index()
    dt_std.columns = ['Data','Ativos','std']

    f_mm = px.line(
            dt_mm,
            x = 'Data',
            y = 'MM',
            color='Ativos'
        )
    f_std = px.line(
            dt_std,
            x = 'Data',
            y = 'std',
            color='Ativos'
        )

    # Criando um MultiPlot no Plotly
    f_mm_traces = []
    f_std_traces = [] 
    for trace in range(len(f_mm['data'])):
        f_mm_traces.append(f_mm['data'][trace])
    for trace in range(len(f_std['data'])):
        f_std_traces.append(f_std['data'][trace])      

    figura = sp.make_subplots(rows=2, cols=1,
                              subplot_titles=('Média Móvel','Desvio Padrão')
                             )
    figura.update_layout({'template':'plotly_white',
                          'height':1100,
                          'width':1000})
    for traces in f_mm_traces:
        figura.append_trace(traces, row=1, col=1)    
    for traces in f_std_traces:
        figura.append_trace(traces, row=2, col=1)  
    
    
    figura.show()
````

````python
analise(df, 'Adj Close',90)
````
