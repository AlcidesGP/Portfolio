# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
df['media'] = df[['p1','p2']].mean(1)
df['evolucao'] = df['p2'] - df['p1']
df.head()
```

```python
def corre(pos1, pos2) -> None:
    
    plt.figure(figsize=(15,5))
    sns.regplot(x = df[pos1],y = df[pos2])
    plt.title('Gráfico de Correlação', fontsize=20)
    plt.show()
    
```

```python
corre('p1','p2')
```

```python
corre('media','evolucao')
```

- Analisando os gráficos gerados, observa-se na primeira imagem que a correlação entre o P1 e P2 é positiva, ou seja, ambos se movem na mesma direção. Já avaliando o segundo gráfico, percebe-se que a média e a diferença tem uma relação praticamente neutra, sendo levemente positiva, mas estando na margem de error.

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
df['soma'] = df['p1'] + df['p2']
```

```python
corre('soma','evolucao')
```

- O resultado de correlação, observado no gráfico acima, se assemelha muito ao do gráfico de correlação entre média e diferença.
