# Tarefa - lgebra Linear

A clula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlaes

a) J vimos que expressar os dados atravs da mdia e diferena  uma transformao linear. Obtenha a mdia e a diferena atravs de uma multiplicao matricial, conforme visto em aula.

b) Calcule a correlao entre os dados da **p1** e **p2**

c) Calcule a correlao entre os dados transformados (mdia e diferena)

d) Comente as correlaes observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
df['media'] = df[['p1','p2']].mean(1)
df['evolucao'] = df['p2'] - df['p1']
df.head()
```

```python
def corre(pos1, pos2) -> None:
    
    plt.figure(figsize=(15,5))
    sns.regplot(x = df[pos1],y = df[pos2])
    plt.title('Grfico de Correlao', fontsize=20)
    plt.show()
    
```

```python
corre('p1','p2')
```

```python
corre('media','evolucao')
```

- Analisando os grficos gerados, observa-se na primeira imagem que a correlao entre o P1 e P2  positiva, ou seja, ambos se movem na mesma direo. J avaliando o segundo grfico, percebe-se que a mdia e a diferena tem uma relao praticamente neutra, sendo levemente positiva, mas estando na margem de error.

## 2. Soma e diferena

Realize agora uma transformao semelhante, mas que ao invs da mdia, d a soma da **p1** e **p2**. Construa o grfico e conclua que diferenas visuais voc observa.

```python
df['soma'] = df['p1'] + df['p2']
```

```python
corre('soma','evolucao')
```

- O resultado de correlao, observado no grfico acima, se assemelha muito ao do grfico de correlao entre mdia e diferena.
