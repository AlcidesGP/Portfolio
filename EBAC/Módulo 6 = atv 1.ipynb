####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
def imagem_barras() -> None:
    """
    Função criada para imprimir a dispersão dos dados qualitativos.    
    """
    
    
    figura = plt.figure(figsize=(15,20)) # Criar imagem e tamanho da mesma.
    figura.suptitle('Gráfico de dispersão dos dados categóricos.', fontsize=30) # Título
    figura.subplots_adjust(wspace=0.5, hspace=0.4) # Ajuste de distância entre os gráficos.
    
    
    tamanho = (3,2) # Quantidade de gráficos (3 linhas e 2 colunas)
    col = 0 # Contador.
    dt = df.copy() # Cópia de segurança.
    atributos = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
    
    
    
    for pos_x in range(3): # Eixo X.
        for pos_y in range(2): # Eixo Y.
            
             # ---------- Configurações            
            nome_coluna =  atributos[col] # Selecionar qual coluna trabalhar.
            pos = (pos_x, pos_y) # Definir a posição da imagem.
            
            if pos_x == 2: # Como o número é ímpar de dados (5), essa lógica "estíca" o último gráfico para 2 colunas.
                ax = plt.subplot2grid(tamanho, pos, colspan=2)
            else:
                ax = plt.subplot2grid(tamanho, pos)
            
                
            # --------- Plot gráfica
            dt[nome_coluna] = dt[nome_coluna].fillna('Não Informado') # Localizar o quão significativo os dados vazios são.
            valores = dt[nome_coluna].value_counts().values # Seleciona qual dados serão trabalhados.
            indices = dt[nome_coluna].value_counts().index.astype(str) # Seleciona qual dados serão trabalhados.
            
            if len(indices[0]) > 10:
                indices = [x.replace('semanas','\nsemanas') for x in indices]
            
            barra = sns.barplot(y = indices,
                                x = valores,
                                palette='magma',
                                orient='h')
            
            # --------- inserir informações no gráfico
            
            for contagem, valor_unidade in zip(range(len(valores)), valores):
            
                porcetagem = (valor_unidade/(valores.sum()))*100
                texto = f" {valor_unidade} | {round(porcetagem,2)}%"
                
                 # Para o gráfico não ficar com a visualização comprometida, valores abaixo de 10% ficam na pos de 1/15 da máxima 
                if porcetagem < 1:
                    text_pos = valores.max()/15 
                else:
                    text_pos = valor_unidade
                
                barra.annotate(texto, xy = (text_pos, contagem), fontsize=10, verticalalignment='center')
            
            # --------- Ajustes de cada gráfico.
            
            barra.set_title(atributos[col], fontsize=20, color='grey') # Título de cada gráfico.           
            barra.spines['top'].set_visible(False) # Desativar as barras da figura.
            barra.spines['right'].set_visible(False)
            barra.spines['left'].set_visible(False)
            barra.tick_params(axis='x', colors='darkgrey') # Trocar a cor dos números do eixo X.
            barra.set(xlabel=f'Frequência das categorias do atributo {nome_coluna}', 
                      ylabel='',
                      xlim = ([0,max(valores)*1.2])) # Aumentar o limite de cada barra em 20%
            
            col += 1  # Adicionar ao contador
            if pos_x == 2: # Como o último gráfico foi esticado, então o break quebra a função para somente mostrar o primeiro gráfico
                break
    
    
    return plt.show()
```

```python
imagem_barras()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
def imagem_barra_mean() -> None:
    figura = plt.figure(figsize=(10,5))
    figura.suptitle('APGAR5 médio por Consulta', fontsize=25, y=1.05)
    
    valores = df.groupby('CONSULTAS')['APGAR5'].mean().values
    indice = df.groupby('CONSULTAS')['APGAR5'].mean().index.astype(str).tolist()
    
    barra = sns.barplot(y = indice, x = valores, orient = 'h', palette = 'viridis')
    
    for contagem, values in zip(range(len(indice)), valores):
        
        porcetagem = (values/(valores.sum()))*100
        texto = f" {round(values,2)} | {round(porcetagem,2)}%"
                
        barra.annotate(texto, xy = (values, contagem), fontsize=10, verticalalignment='center')
    barra.spines['top'].set_visible(False) # Desativar as barras da figura.
    barra.spines['right'].set_visible(False)
    barra.set(xlabel=f'Média por Consulta do APGAR5', 
              ylabel='Classes de Consulta',
              xlim=([0,max(valores)*1.2]))
    
    
    
    
    
    return plt.show()
```

```python
imagem_barra_mean()
```
